```python
class NodoDoble:
    def __init__( self, value, siguiente = None, previo = None ):
        self.data = value 
        self.next = siguiente 
        self.prev = previo 

class DoubleLinkedList:
    def __init__( self  ):
        self.head = None
        self.tail = None
        self.size = 0

    def get_size( self ):
        return self.size

    def is_empty( self ): # true si la lista está vacía
        return self.size == 0

    def append( self , value ):
        if self.is_empty():
            self.head = NodoDoble( value)
            self.tail = self.head
        else:
            self.tail.next = NodoDoble(value , None , self.tail )
            self.tail = self.tail.next
        self.size += 1

    def find_from_tail( self, value ):
        contador = 1
        if self.existe( value ):
            curr_node = self.tail
            while curr_node.data != value:
                curr_node = curr_node.prev
                contador += 1
            print(f"El dato | {value} | se encuentra en la posición: {contador} \n(Donde TAIL es el dato No. 1)")
        
        else:
            print("El valor solicitado no existe")

    def find_from_head( self, value ):
        contador = 1
        if self.existe( value ):
            curr_node = self.head
            while curr_node.data != value:
                curr_node = curr_node.next
                contador += 1
            print(f"El dato | {value} | se encuentra en la posición: {contador} \n(Donde HEAD es el dato No. 1)")
        
        else:
            print("El valor solicitado no existe")

    def remove_from_tail( self , value  ):
        if self.existe( value ):
            curr_node = self.tail #Entrando por TAIL
            while curr_node.data != value:
                curr_node = curr_node.prev

            if curr_node.data == self.head.data: #Si el elemento a eliminar es el head
                if self.head.next != None: #Si existe el elemento siguiente a head, este será el nuevo head
                    self.head = self.head.next
                    self.head.prev = None
                    print(f"Se eliminó el dato: | {curr_node.data} | que era el elemento HEAD")
                    print(f"Nuevo HEAD: | {self.head.data} |")
                else: #Si solamente existía head simplemente se elimina
                    self.head = None
                    print(f"Se eliminó el dato: | {curr_node.data} | , la lista doble ligada está vacía ahora")
            
            elif curr_node.data == self.tail.data: #Si el elemento a eliminar es tail, el penúltimo se vuelve el nuevo tail 
                self.tail = self.tail.prev
                self.tail.next = None
                print(f"Se eliminó el dato: | {curr_node.data} | que era el elemento TAIL")
                print(f"Nuevo TAIL: | {self.tail.data} |")

            elif curr_node.data == value: #Si es un valor intermedio 
                print("Se eliminó el elemento: ",curr_node.data)
                curr_node.prev.next = curr_node.next
                curr_node.next.prev = curr_node.prev
            self.size -= 1

        else:
            print("No existe el elemento a remover")

    def remove_from_head( self , value  ):
        if self.existe( value ):
            curr_node = self.head #Entrando por HEAD
            while curr_node.data != value:
                curr_node = curr_node.next

            if curr_node.data == self.head.data: #Si el elemento a eliminar es el head
                if self.head.next != None: #Si existe el elemento siguiente a head, este será el nuevo head
                    self.head = self.head.next
                    self.head.prev = None
                    print(f"Se eliminó el dato: | {curr_node.data} | que era el elemento HEAD")
                    print(f"Nuevo HEAD: | {self.head.data} |")
                else: #Si solamente existía head simplemente se elimina
                    self.head = None
                    print(f"Se eliminó el dato: | {curr_node.data} | , la lista doble ligada está vacía ahora")
            
            elif curr_node.data == self.tail.data: #Si el elemento a eliminar es tail, el penúltimo se vuelve el nuevo tail 
                self.tail = self.tail.prev
                self.tail.next = None
                print(f"Se eliminó el dato: | {curr_node.data} | que era el elemento TAIL")
                print(f"Nuevo TAIL: | {self.tail.data} |")

            elif curr_node.data == value: #Si es un valor intermedio 
                print("Se eliminó el elemento: ",curr_node.data)
                curr_node.prev.next = curr_node.next
                curr_node.next.prev = curr_node.prev
            self.size -= 1

        else:
            print("No existe el elemento a remover")

    def transversal( self ):
        curr_node = self.head
        while curr_node != None :
            print(f"| {curr_node.data} | --> " , end="")
            curr_node = curr_node.next
        print("")

    def reverse_transversal( self ):
        curr_node = self.tail
        while curr_node != None :
            print(f"| {curr_node.data} | --> " , end="")
            curr_node = curr_node.prev
        print("")

#NOTA AGREGUÉ UN NUEVO MÉTODO QUE ME PARECIO ÚTIL LLAMADO "EXISTE", SIRVE PARA SABER SI DETERMINADO ELEMENTO EXISTE EN NUESTRA DLL O NO
    def existe( self, value ):
        curr_node = self.head
        continuar = True
        existe = False

        while continuar != False:
            if curr_node.data == value: #Si curr_node contiene el dato requrido salimos del while
                continuar = False
                existe = True
            else: #De lo contrarios, seguimos recorriendo la estructura aumentado el contador
                curr_node = curr_node.next
            if curr_node == None: #Si hemos recorrido ya todos los datos y no hubo coincidencias, también salimos del while
                continuar = False
                existe = False

        return existe
            
```

```python
print("------ PRUEBAS DOBLE LINKED LIST ------")
dll1 = DoubleLinkedList()
print("\n---> Prueba get_size, is_empty y append:")
print(f"Lista doblemente ligada vacía: {dll1.is_empty()}")
print(f"No. de elementos en la lista inicial: {dll1.get_size()}")
print("\nSe agregarán los datos: 10, 20, 30, 40 y 50")
dll1.append(10)
dll1.append(20)
dll1.append(30)
dll1.append(40)
dll1.append(50)
print(f"Lista doblemente ligada vacía: {dll1.is_empty()}")
print(f"No. de elementos en la lista inicial: {dll1.get_size()}")

print("\n---> Prueba transversal y reverse_transversal:")
dll1.transversal()
dll1.reverse_transversal()

print("\n---> Prueba find_from_tail y find_from_head:")
print("Buscar la posición del número 20 a partir de la posición de TAIL:")
dll1.find_from_tail(20)
print("Buscar la posición del número 50 a partir de la posición de HEAD:")
dll1.find_from_head(50)

print("\n---> Prueba remove_from_tail y remove_from_head:")
print("Eliminar el valor 20 a partir de TAIL")
dll1.remove_from_tail(20)
dll1.transversal()
print("Eliminar el valor 30 a partir de TAIL:")
dll1.remove_from_tail(30)
dll1.transversal()
print("\n- PRUEBAS ADICIONALES -")
print("De una lista con un solo elemento eliminarlo con ambos métodos:")
dll2 = DoubleLinkedList()
dll2.append(10)
dll2.transversal()
dll2.remove_from_tail(10)
print(f"Tamaño de la lista: {dll2.get_size()}")
print(f"¿Está vacía? R= {dll2.is_empty()}")
dll2.append(20)
dll2.transversal()
dll2.remove_from_head(20)
print(f"Tamaño de la lista: {dll2.get_size()}")
print(f"¿Está vacía? R= {dll2.is_empty()}")

print("\nDe una lista eliminar HEAD con remove_from_head y a TAIL con remove_from_tail:")
dll2.append(1)
dll2.append(2)
dll2.append(3)
dll2.append(4)
dll2.append(5)
dll2.transversal()
dll2.remove_from_head(1)
dll2.remove_from_tail(5)
dll2.transversal()

print("\nFinalmente, intentar eliminar el numero 100 de la lista anterior")
dll2.remove_from_head(100)
dll2.remove_from_tail(100)
```
